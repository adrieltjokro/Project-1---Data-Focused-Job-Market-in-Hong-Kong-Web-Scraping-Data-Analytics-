### 1. Import the necessary libraries

```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import time
import pandas as pd
import numpy as np
import os
```

### 2. Do the extraction

```python
### Build the dictionary to record the data for each job posting ###

d = {
    'overall_score':[],
    'company':[],
    'title':[],
    'location':[],
    'EmpBasicInfo':[],
    'employerStats':[],
    'JobDescriptionContainer':[],
    'salary':[]
}

### Based on the number of pages seen for a particular role, create the loop to scrape all pages ###
### As of 18 June 2021 at 9pm, there are 230 job postings for Data Analysts in 8 pages ###

for i in range(1,10):
    driver = webdriver.Chrome('./chromedriver')
    url = f"https://www.glassdoor.com.hk/Job/data-analyst-jobs-SRCH_KO0,12_IP{i}.htm"
    driver.get(url)
    
    time.sleep(5)
    
    try:
        driver.find_element_by_class_name("modal_closeIcon").click() # SVGInline-svg modal_closeIcon-svg
    except:
        print("pop up issue")
    finally:
        
        for i in driver.find_elements_by_id('JobResults'):
            for j in i.find_elements_by_id('MainCol'):
                for k in j.find_elements_by_tag_name('li'):
                    try:
                        driver.find_element_by_class_name("modal_closeIcon").click() #SVGInline modal_closeIcon
                    except Exception as e:
                        print(e)

                    try:
                        d['overall_score'].append(k.find_element_by_class_name("css-19pjha7.e1cjmv6j1").text)
                    except:
                        d['overall_score'].append('NA')
                    try:
                        d['company'].append(k.find_element_by_class_name("css-l2wjgv.e1n63ojh0.jobLink").text)
                    except:
                        d['company'].append('NA')
                    try:
                        d['title'].append(k.find_element_by_class_name("jobLink.css-1rd3saf.eigr9kq2").text)
                    except:
                        d['title'].append('NA')
                    try:
                        d['location'].append(k.find_element_by_class_name("pr-xxsm.css-1ndif2q.e1rrn5ka0").text)
                    except:
                        d['location'].append('NA')
                    try:
                        d['salary'].append(k.find_element_by_class_name("css-1imh2hq.e1wijj242").text)
                    except:
                        d['salary'].append('NA')

                    try:
                        k.find_element_by_class_name("css-l2wjgv.e1n63ojh0.jobLink").click()
                        time.sleep(5)
                        try:
                            driver.find_element_by_class_name("SVGInline-svg modal_closeIcon-svg").click()
                            k.find_element_by_class_name("css-l2wjgv.e1n63ojh0.jobLink").click()
                        except:
                            time.sleep(1)
                    except:
                        pass 

                    
                    try:
                        d['EmpBasicInfo'].append(driver.find_element_by_id('EmpBasicInfo').text)
                    except:
                        d['EmpBasicInfo'].append('NA')
                    try:
                        d['employerStats'].append(driver.find_element_by_id('employerStats').text)
                    except:
                        d['employerStats'].append('NA')

                    try:
                        d['JobDescriptionContainer'].append(driver.find_element_by_id('JobDescriptionContainer').text)
                    except:
                        d['JobDescriptionContainer'].append('NA')

    driver.close()
```

### 3. Clean up on a macro basis

#### Firstly drop any duplicate rows

```python
df = pd.DataFrame(d.copy()).drop_duplicates()
df
```

#### Then drop any rows that do not have any company names
This is to ensure that all of our data points are of high-quality. Our group believes that high-quality data points would possess company names, even if said companies are recruiting companies.

```python
df = df[df['company'].str.contains("NA")==False]
df
```

#### Overall score values are not extracted properly, so let's extract those values from the employerStats

```python
### Gather all elements from 'employerStats' column ###

ratings = []

for i in df['employerStats']:
    ratings.append(i)

### Then extract the first 3 characters ###

import re

new_overall_score = []
    
for j in ratings:
    if j == 'NA':
        new_overall_score.append('0.0')
    else:
        new_overall_score.append(j[0:3])

### Now we can drop the old 'overall_score' column and add in the 'new_overall_score' column ###
### Also change the 'new_overall_score' column to 'float64' ###

df['new_overall_score'] = new_overall_score
df1 = df.drop(columns='overall_score')
df1 = df1.astype({'new_overall_score': 'float64'})
df1
```

### 5. Clean-up the data as necessary

Based on examples above, only 'EmpBasicInfo' and 'salary' needs to be cleaned up. The rest contains unnecessary data for our analyses.

#### 'EmpBasicInfo'

```python
### Create the lists that will separate keys and values into their respective pairs ###

variables = []
data = []

for i in list(df1['EmpBasicInfo']):
    variables.append((i.split('\n'))[1::2][:-1])
    data.append((i.split('\n'))[2::2])
```

```python
### We know that there are 177 unique entries, so we use range(178). ###
### For each variable in the 'EmpBasicInfo', combine them with the respective value. ###
### Let's combine all in a list of dictionaries. ###

basic_info = []

for i in range(178):
    for j in list(zip(variables, data)):
        basic_info.append(dict(zip(j[0], j[1])))
        
pd.DataFrame(basic_info)
```

#### 'salary'

```python
### First step is to extract all salary information into a list (including the Glassdoor estimates text, etc.) ###

salary = []

for j in df1['salary']:
    salary.append(j)

### Afterwards, we separate the figures into each respective interquartiles ###
    
import re
salary_range = []
    
for k in salary:
    if k == 'NA':
        salary_range.append('NA')
    else:
        salary_range.append(re.findall(r'\d+',k))

### Those jobs who don't have salary figures will be recorded as 'NA' ##
### Those jobs with only one salary figure will record this figure as both lower and upper range ###
### Those jobs with two salary figures will have the first one as lower range and second one as upper range ###
        
salary_lower = []
salary_upper = []

for l in salary_range:
    if l == 'NA':
        salary_lower.append('NA')
        salary_upper.append('NA')
    elif len(l) == 1:
        salary_lower.append(l[0])
        salary_upper.append(l[0])
    else:
        salary_lower.append(l[0])
        salary_upper.append(l[1])

### Finally, make a dictionary to make into a dataframe ###

dict_salary = {'salary_lower': salary_lower, 'salary_upper': salary_upper}
dict_salary
df_salary = pd.DataFrame(dict_salary)

### As salary figures are quantative values, change them into integers ###
### Because there are so many NAs, in order to do salary analysis, we have to remove those ###
### So let's combine this dataframe with the original dataframe first ###
# df_salary = df_salary['salary_lower'].astype(str).astype(int)
```
